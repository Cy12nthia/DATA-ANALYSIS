## ðŸ“„ Dataset Column Descriptions

1. **Row ID**  - A unique identifier for each row/record in the dataset. Mainly used for indexing, not analysis.

2. **Order ID**  - A unique identifier for each customer order. One order can contain multiple products.

3. **Order Date**  -  The date when the customer placed the order. Used for time-based analysis such as trends and seasonality.

4. **Ship Date**  - The date when the order was shipped to the customer.

5. **Ship Mode**  - The shipping method used (Same Day, First Class, Second Class, Standard Class).

6. **Customer ID**  - A unique identifier assigned to each customer.

7. **Customer Name**  - The full name of the customer who placed the order.

8. **Segment**  - Customer classification (Consumer, Corporate, Home Office). Useful for segmentation analysis.

9. **Country**  - The country where the order was placed (mostly United States in this dataset).

10. **City**  - The city where the order was delivered.

11. **State**  - The state where the order was delivered.

12. **Postal Code**   - The postal/ZIP code of the delivery location.

13. **Region**  - The sales region (East, West, Central, South). Useful for regional performance analysis.

14. **Product ID**  - A unique identifier for each product.

15. **Category**  - The high-level product category (Furniture, Office Supplies, Technology).

16. **Sub-Category**  - A more detailed product classification within a category (e.g., Chairs, Phones, Storage).

17. **Product Name**  - The name or description of the product sold.

18. **Sales**  - The total sales revenue generated from the order (after discounts).

19. **Quantity**  - The number of units of the product sold in the order.

20. **Discount**  - The discount applied to the product (as a fraction, e.g., 0.2 = 20%).

21. **Profit**  - The net profit earned from the sale after subtracting costs and discounts.
