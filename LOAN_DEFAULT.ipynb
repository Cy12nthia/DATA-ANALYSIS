# 5. FEATURE ENGINEERING

```python
# Encode the categorical variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['MaritalStatus'] = encoder.fit_transform(data['MaritalStatus'])
data['Education'] = encoder.fit_transform(data['Education'])
data['EmploymentType'] = encoder.fit_transform(data['EmploymentType'])
data['LoanPurpose'] = encoder.fit_transform(data['LoanPurpose'])
data['HasDependents'] = encoder.fit_transform(data['HasMortgage'])
data['HasMortgage'] = encoder.fit_transform(data['HasMortgage'])
data['HasCoSigner'] = encoder.fit_transform(data['HasCoSigner'])
```

```python
# define the dependent and independent variables
X = data.drop('Default', axis =1)
y = data['Default']
```

```python
# split into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.4 , random_state = 42)
```

```python
# normalize the  data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

Normalize the x_test and X_train to be standardized




```python
# fit the machine learning model
#  since the output is in ones and zero we use logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# evaluate the model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

- The model has an accuracy score of 88%
- The model is able to correctly identify loan defaults at 89%  
- The model is able to predict the default will not happen at 61% accurately
