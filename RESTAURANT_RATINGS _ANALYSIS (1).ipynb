## 3. MULTIVARIATE ANALYSIS 

```python
# Check the distibution per two values 
sns.scatterplot(x='votes', y='rate', data=data)
plt.title("Votes vs Rating")
plt.show()
```

- From the table higher rate is slightly influenced by the number of votes 

```python
# check the correlation between votes and cost 
sns.scatterplot(x='votes', y='approx_cost(for two people)', data=data)
plt.title("Votes vs cost")
plt.show()
```

- From the plot votes increases slightly as the cost increases

```python
# check the corelation between cost and votes 
data[['approx_cost(for two people)', 'votes']].corr()
```

- The relation between cost and votes is a weak positive correlation This means the votes  increases slightly as the cost increases .

```python
# ENCODE THE RATES TO numeric 
data['rate'] = data['rate'].astype(str)
data['rate'] = data['rate'].str.split('/').str[0]
data['rate'] = data['rate'].replace(['NEW', '-', 'nan'], None)
data['rate'] = data['rate'].astype(float)
```

```python
data[['rate', 'votes']].corr()
```

. There is a moderate positive relationship between votes and ratings (correlation â‰ˆ 0.4), suggesting that restaurants with higher customer engagement tend to receive higher ratings. However, the relationship is not strong, indicating that other factors also influence restaurant ratings.

```python
# the corre;ation between cost and ratings 
data[['rate', 'approx_cost(for two people)']].corr()
```

- There is a weak positive relationship between cost and ratings (correlation â‰ˆ 0.2), indicating that higher-priced restaurants tend to have slightly higher ratings. However, cost alone is a poor predictor of restaurant ratings.

```python
# votes vs restaurants type 
grouped_data = data.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of restaurant')
plt.ylabel('Votes')
```

- Dining restaurants have higher votes than others 

```python
# rates  vs restaurants type 
grouped_data = data.groupby('listed_in(type)')['rate'].sum()
result = pd.DataFrame({'rate': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of restaurant')
plt.ylabel('rates')
```

- dining restaurants have higher ratings compared to others
